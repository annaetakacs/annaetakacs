```python
import pandas as pd
import re
import os
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# my_stem = PorterStemmer()
my_stem = WordNetLemmatizer()

stop_words = set(stopwords.words('english'))
stop_words.add('e')

the_path = '/Users/annatakacs/Documents/GitHub/GR5067_NLP/topics/'
the_dirs = os.listdir(the_path)
print (the_dirs)

def clean_data(var, the_path, the_dirs):
        f = open(the_path + dir_name + '/' + var, "r",
                 encoding='ISO-8859-1')
        tmp_read = re.sub('[^a-zA-Z]+', ' ', str(f.read()))
        f.close()
        tmp_read = [word.lower() for word in tmp_read.split() if word not in stop_words]
        # Stem the document
        # tmp_read = [my_stem.stem(word) for word in tmp_read] # Stem each word
        tmp_read = [my_stem.lemmatize(word) for word in tmp_read]
        tmp_read = ' '.join(tmp_read)
        
        return tmp_read


# HW for the next assignment: turn everything into lambda function     

the_df = pd.DataFrame()
for dir_name in the_dirs:
    the_filenames = os.listdir(the_path + dir_name)
    for word in the_filenames:
        tmp_read = clean_data(word, the_path, the_dirs)
        tmp = pd.DataFrame([tmp_read], columns=['body'])
        tmp['label'] = dir_name
        the_df = the_df.append(tmp, ignore_index=True)
        
the_df.head()
```
